```python
# CS770 Assignment 2 • BMI Classification

# Setup
!pip -q install imbalanced-learn==0.12.3

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report
)

from imblearn.over_sampling import SMOTE

# Reproducibility
SEED = 42
np.random.seed(SEED)

# Folders for outputs
FIG_DIR = Path("figures"); FIG_DIR.mkdir(exist_ok=True)
OUT_DIR = Path("outputs"); OUT_DIR.mkdir(exist_ok=True)

# Helpful mapping
bmi_map = {
    0: "Extremely Weak",
    1: "Weak",
    2: "Normal",
    3: "Overweight",
    4: "Obesity",
    5: "Extreme Obesity"
}

# Question 1. Data Exploration and Cleanup
#   1.1 Load data
#   1.2 Inspect structure and missing values
#   1.3 Remove duplicates and implausible rows

# 1.1 Load data
# Load from GitHub since the data is uploaded in GitHub
bmi_url = "https://raw.githubusercontent.com/TheRufael/CS770-Assignments/main/bmi.csv"
df = pd.read_csv(bmi_url)

print("Head")
display(df.head())
print("Shape:", df.shape)
print("Columns:", list(df.columns))

# 1.2 Inspect structure and missing values
print("\nInfo")
print(df.info())
print("\nDescribe")
display(df.describe())
print("\nMissing values per column")
print(df.isnull().sum())

# 1.3 Cleanup
before = len(df)
df = df.drop_duplicates()
after = len(df)
print("\nDropped duplicates:", before - after)

# Keep plausible ranges to remove obvious anomalies
mask = (
    df["Height"].between(120, 220) &
    df["Weight"].between(30, 200) &
    df["Gender"].isin(["Male", "Female"]) &
    df["Index"].isin([0, 1, 2, 3, 4, 5])
)
removed = len(df) - mask.sum()
df = df[mask].copy()
print("Removed implausible rows:", removed)
df["IndexName"] = df["Index"].map(bmi_map)

# Question 2. Exploratory Data Analysis
#   2.1 Class distribution
#   2.2 Gender distribution
#   2.3 Height and Weight distributions
#   2.4 Gender-specific distributions
#   2.5 Correlation and scatter

sns.set_theme(style="whitegrid")
order_names = [bmi_map[i] for i in sorted(bmi_map)]

# 2.1 Class distribution
plt.figure(figsize=(7,4))
sns.countplot(x="IndexName", data=df, order=order_names)
plt.xticks(rotation=25)
plt.title("BMI category distribution")
plt.tight_layout()
plt.savefig(FIG_DIR / "Q2_Fig1_BMI_Category_Distribution.png", dpi=200)
plt.show()

# 2.2 Gender distribution
plt.figure(figsize=(5,4))
sns.countplot(x="Gender", data=df)
plt.title("Gender distribution")
plt.tight_layout()
plt.savefig(FIG_DIR / "Q2_Fig2_Gender_Distribution.png", dpi=200)
plt.show()

# 2.3 Height and Weight distributions
sns.histplot(df["Height"], bins=24, kde=True)
plt.title("Height distribution")
plt.tight_layout()
plt.savefig(FIG_DIR / "Q2_Fig3_Height_Distribution.png", dpi=200)
plt.show()

sns.histplot(df["Weight"], bins=24, kde=True)
plt.title("Weight distribution")
plt.tight_layout()
plt.savefig(FIG_DIR / "Q2_Fig4_Weight_Distribution.png", dpi=200)
plt.show()

# 2.4 Gender-specific distributions
plt.figure(figsize=(7,4))
sns.histplot(data=df, x="Height", hue="Gender", bins=24, kde=True, element="step")
plt.title("Height by gender")
plt.tight_layout()
plt.savefig(FIG_DIR / "Q2_Fig5_Height_By_Gender.png", dpi=200)
plt.show()

plt.figure(figsize=(7,4))
sns.histplot(data=df, x="Weight", hue="Gender", bins=24, kde=True, element="step")
plt.title("Weight by gender")
plt.tight_layout()
plt.savefig(FIG_DIR / "Q2_Fig6_Weight_By_Gender.png", dpi=200)
plt.show()

plt.figure(figsize=(8,4))
sns.countplot(data=df, x="IndexName", hue="Gender", order=order_names)
plt.xticks(rotation=25)
plt.title("BMI category by gender")
plt.tight_layout()
plt.savefig(FIG_DIR / "Q2_Fig7_BMI_By_Gender_Counts.png", dpi=200)
plt.show()

# 2.5 Correlation and scatter
corr = df[["Height","Weight","Index"]].corr()
plt.figure(figsize=(4,3))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="Blues")
plt.title("Correlation heatmap")
plt.tight_layout()
plt.savefig(FIG_DIR / "Q2_Fig8_Correlation_Heatmap.png", dpi=200)
plt.show()

plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="Height", y="Weight", hue="IndexName", s=35, alpha=0.85)
plt.title("Height vs Weight by BMI")
plt.tight_layout()
plt.savefig(FIG_DIR / "Q2_Fig9_Scatter_Height_Weight_BMI.png", dpi=200)
plt.show()

# Question 3. Data Preprocessing
#   3.1 Encode Gender
#   3.2 Train test split with stratification by gender and class
#   3.3 Handle imbalance with SMOTE on training only
#   3.4 Scale features

# 3.1 Encode Gender
X_all = pd.get_dummies(df[["Gender","Height","Weight"]], drop_first=True)
y_all = df["Index"]

# 3.2 Stratified split by combined gender and class
strata = df["Gender"].astype(str) + "_" + df["Index"].astype(str)
X_train, X_test, y_train, y_test = train_test_split(
    X_all, y_all, test_size=0.2, random_state=SEED, stratify=strata
)

# 3.3 SMOTE on training only
smote = SMOTE(random_state=SEED)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# 3.4 Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_res)
X_test_scaled  = scaler.transform(X_test)

# Question 4. Model Training and Prediction
#   4.1 Baseline LR, SVM, KNN
#   4.2 Hyperparameter tuning for LR, SVM, KNN
#   4.3 Save metrics and confusion matrices

def evaluate_model(y_true, y_pred, model_name, fig_prefix):
    print(f"\nModel. {model_name}")
    acc = accuracy_score(y_true, y_pred)
    pre = precision_score(y_true, y_pred, average="weighted", zero_division=0)
    rec = recall_score(y_true, y_pred, average="weighted", zero_division=0)
    f1  = f1_score(y_true, y_pred, average="weighted", zero_division=0)
    print("Accuracy:", acc)
    print("Precision weighted:", pre)
    print("Recall weighted:", rec)
    print("F1 weighted:", f1)
    print("\nClassification report")
    print(classification_report(
        y_true, y_pred, target_names=[bmi_map[i] for i in sorted(bmi_map)], zero_division=0
    ))

    cm = confusion_matrix(y_true, y_pred, labels=sorted(bmi_map))
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=[bmi_map[i] for i in sorted(bmi_map)],
                yticklabels=[bmi_map[i] for i in sorted(bmi_map)])
    plt.title(f"Confusion Matrix. {model_name}")
    plt.tight_layout()
    plt.savefig(FIG_DIR / f"{fig_prefix}_ConfusionMatrix.png", dpi=200)
    plt.show()

    return {"Accuracy": acc, "Precision": pre, "Recall": rec, "F1": f1}

# 4.1 Baseline models
lr_baseline  = LogisticRegression(max_iter=500, random_state=SEED)
svm_baseline = SVC(random_state=SEED)
knn_baseline = KNeighborsClassifier()

gen_base_results = {}

for est, name, prefix in [
    (lr_baseline,  "Logistic Regression Baseline", "Q4_General_LR_Baseline"),
    (svm_baseline, "SVM Baseline", "Q4_General_SVM_Baseline"),
    (knn_baseline, "KNN Baseline", "Q4_General_KNN_Baseline"),
]:
    est.fit(X_train_scaled, y_train_res)
    y_pred = est.predict(X_test_scaled)
    gen_base_results[name] = evaluate_model(y_test, y_pred, name, prefix)

pd.DataFrame(gen_base_results).T.to_csv(OUT_DIR / "Q4_General_Baseline_Metrics.csv")

# 4.2 Hyperparameter tuning
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

lr = LogisticRegression(max_iter=500, random_state=SEED)
lr_grid = {"C":[0.1,1,10], "penalty":["l2"]}

svm = SVC(random_state=SEED)
svm_grid = {
    "kernel":["rbf","linear","poly"],
    "C":[0.5,1,5,10],
    "gamma":["scale","auto"],
    "degree":[2,3]
}

knn = KNeighborsClassifier()
knn_grid = {
    "n_neighbors":[3,5,7,9,11],
    "weights":["uniform","distance"],
    "p":[1,2]
}

def tune_and_eval(estimator, grid, label, prefix):
    gs = GridSearchCV(estimator=estimator, param_grid=grid, cv=cv,
                      n_jobs=-1, scoring="accuracy", refit=True, verbose=0)
    gs.fit(X_train_scaled, y_train_res)
    best = gs.best_estimator_
    y_pred = best.predict(X_test_scaled)
    print(f"\nBest params for {label}")
    print(gs.best_params_)
    metrics = evaluate_model(y_test, y_pred, f"{label} Tuned + SMOTE", prefix)
    return metrics, gs.best_params_

gen_lr_metrics, gen_lr_params   = tune_and_eval(lr,  lr_grid,  "Logistic Regression", "Q4_General_LR_Tuned")
gen_svm_metrics, gen_svm_params = tune_and_eval(svm, svm_grid, "SVM",                 "Q4_General_SVM_Tuned")
gen_knn_metrics, gen_knn_params = tune_and_eval(knn, knn_grid, "KNN",                 "Q4_General_KNN_Tuned")

pd.DataFrame({"LR":gen_lr_metrics,"SVM":gen_svm_metrics,"KNN":gen_knn_metrics}).T.to_csv(OUT_DIR / "Q4_General_Tuned_SMOTE_Metrics.csv")
pd.DataFrame({"LR":gen_lr_params,"SVM":gen_svm_params,"KNN":gen_knn_params}).to_csv(OUT_DIR / "Q4_General_Best_Params.csv")

# Additional Visualization: Decision Boundaries for LR, SVM, and KNN

def plot_decision_boundaries(X, y, models, title_prefix):
    # Select only Height and Weight for plotting
    X_plot = X[["Height", "Weight"]]

    # Determine grid limits
    h_min, h_max = X_plot["Height"].min() - 5, X_plot["Height"].max() + 5
    w_min, w_max = X_plot["Weight"].min() - 5, X_plot["Weight"].max() + 5
    xx, yy = np.meshgrid(np.arange(h_min, h_max, 1),
                         np.arange(w_min, w_max, 1))

    # Create a DataFrame for grid points
    grid_points = pd.DataFrame(np.c_[xx.ravel(), yy.ravel()], columns=["Height", "Weight"])

    # If Gender is a feature, assume a default value (e.g., 0 for visualization purposes)
    # This is a simplification for visualization and might not reflect real-world scenarios accurately
    if "Gender_Male" in X.columns:
        grid_points["Gender_Male"] = 0 # You can change this to 1 for 'Male' or 0.5 for a midpoint


    # Scale the grid points using the same scaler trained on the original data
    grid_points_scaled = scaler.transform(grid_points)

    # Plot decision boundaries for each model
    for model_name, model in models.items():
        plt.figure(figsize=(8, 6))

        # Predict on the grid points
        Z = model.predict(grid_points_scaled)
        Z = Z.reshape(xx.shape)

        # Plot contour lines or shaded regions
        plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdYlBu)
        plt.scatter(X_plot["Height"], X_plot["Weight"], c=y, s=20, edgecolor='k', cmap=plt.cm.RdYlBu)

        plt.xlabel("Height")
        plt.ylabel("Weight")
        plt.title(f"{title_prefix}: {model_name} Decision Boundaries")
        plt.tight_layout()
        plt.savefig(FIG_DIR / f"{title_prefix.replace(' ', '_')}_{model_name.replace(' ', '_')}_DecisionBoundaries.png", dpi=200)
        plt.show()

# Train models using the combined data (Height, Weight, Gender) as done in Question 4
lr_tuned_viz = LogisticRegression(max_iter=500, random_state=SEED, **gen_lr_params)
svm_tuned_viz = SVC(random_state=SEED, **gen_svm_params)
knn_tuned_viz = KNeighborsClassifier(**gen_knn_params)

models_viz = {
    "Logistic Regression": lr_tuned_viz.fit(X_train_scaled, y_train_res),
    "SVM": svm_tuned_viz.fit(X_train_scaled, y_train_res),
    "KNN": knn_tuned_viz.fit(X_train_scaled, y_train_res)
}

# Plot decision boundaries for the combined data
plot_decision_boundaries(X_all, y_all, models_viz, "Combined Data")


# Question 5. Model Evaluation and Comparison
#   5.1 Compare baseline vs tuned for Combined set
#   5.2 Visualize F1 for Combined models
#   5.3 Save summary tables

gen_base_df  = pd.DataFrame(gen_base_results).T
gen_tuned_df = pd.DataFrame({"LR":gen_lr_metrics,"SVM":gen_svm_metrics,"KNN":gen_knn_metrics}).T

print("\nCombined set. Baseline metrics")
display(gen_base_df)

print("\nCombined set. Tuned + SMOTE metrics")
display(gen_tuned_df)

plt.figure(figsize=(7,5))
plot_df = gen_tuned_df.reset_index().rename(columns={"index":"Model"})
sns.barplot(data=plot_df, x="Model", y="F1")
plt.ylim(0,1)
plt.title("Combined F1 for tuned models")
plt.tight_layout()
plt.savefig(FIG_DIR / "Q5_Fig1_Combined_F1_Tuned.png", dpi=200)
plt.show()

gen_base_df.to_csv(OUT_DIR / "Q5_Combined_Baseline_Metrics.csv")
gen_tuned_df.to_csv(OUT_DIR / "Q5_Combined_Tuned_Metrics.csv")

# Question 6. Gender-Based Prediction Analysis
#   6.1 Train tuned models on Male only
#   6.2 Train tuned models on Female only
#   6.3 Build final comparison table with rows SVM, KNN, LR and columns Combined, Male, Female
#   6.4 Optional. Learning rate test using SGDClassifier

from imblearn.over_sampling import SMOTE, RandomOverSampler
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# 6.1 and 6.2 — robust gender block with adaptive SMOTE, CV, and K for KNN
def run_gender_block(gender_label):
    sub = df[df["Gender"] == gender_label].copy()
    Xg = sub[["Height","Weight"]]  # numeric only
    yg = sub["Index"]

    # Split
    Xg_train, Xg_test, yg_train, yg_test = train_test_split(
        Xg, yg, test_size=0.2, random_state=SEED, stratify=yg
    )

    # Choose resampling method safely
    min_class_train = yg_train.value_counts().min()
    if min_class_train is None or min_class_train < 2:
        print(f"[{gender_label}] Too few samples for SMOTE in at least one class. Using RandomOverSampler.")
        ros = RandomOverSampler(random_state=SEED)
        Xg_train_res, yg_train_res = ros.fit_resample(Xg_train, yg_train)
    else:
        smote_k = max(1, min(5, min_class_train - 1))
        try:
            sm = SMOTE(random_state=SEED, k_neighbors=smote_k)
            Xg_train_res, yg_train_res = sm.fit_resample(Xg_train, yg_train)
        except ValueError:
            print(f"[{gender_label}] SMOTE failed with k={smote_k}. Falling back to RandomOverSampler.")
            ros = RandomOverSampler(random_state=SEED)
            Xg_train_res, yg_train_res = ros.fit_resample(Xg_train, yg_train)

    # Scale
    gscaler = StandardScaler()
    Xg_train_scaled = gscaler.fit_transform(Xg_train_res)
    Xg_test_scaled  = gscaler.transform(Xg_test)

    # Adaptive CV folds for small class counts
    min_class_after = pd.Series(yg_train_res).value_counts().min()
    cv_splits = max(2, min(5, int(min_class_after)))
    cv_local = StratifiedKFold(n_splits=cv_splits, shuffle=True, random_state=SEED)

    # Models and grids
    lr_g   = LogisticRegression(max_iter=500, random_state=SEED)
    svm_g  = SVC(random_state=SEED)
    knn_g  = KNeighborsClassifier()

    lr_grid  = {"C":[0.1,1,10], "penalty":["l2"]}
    svm_grid = {"kernel":["rbf","linear","poly"], "C":[0.5,1,5,10],
                "gamma":["scale","auto"], "degree":[2,3]}

    # KNN grid trimmed so n_neighbors <= n_samples_fit - 1 and prefer odd ks
    n_fit = Xg_train_res.shape[0]
    max_k_allowed = max(1, n_fit - 1)
    candidate_ks = [k for k in [3,5,7,9,11] if k <= max_k_allowed]
    if not candidate_ks:
        k_use = max_k_allowed if max_k_allowed % 2 == 1 else max(max_k_allowed - 1, 1)
        candidate_ks = [k_use]
    knn_grid = {"n_neighbors": candidate_ks, "weights":["uniform","distance"], "p":[1,2]}

    def tune(est, grid, name, prefix):
        gs = GridSearchCV(estimator=est, param_grid=grid, cv=cv_local,
                          n_jobs=-1, scoring="accuracy", refit=True, verbose=0)
        gs.fit(Xg_train_scaled, yg_train_res)
        best = gs.best_estimator_
        y_pred = best.predict(Xg_test_scaled)
        print(f"\n{name} • {gender_label} best params")
        print(gs.best_params_)
        metrics = evaluate_model(yg_test, y_pred, f"{name} • {gender_label} Tuned + Resampled", prefix)
        return metrics

    lr_m  = tune(lr_g,  lr_grid,  "Logistic Regression", f"Q6_{gender_label}_LR_Tuned")
    svm_m = tune(svm_g, svm_grid, "SVM",                 f"Q6_{gender_label}_SVM_Tuned")
    knn_m = tune(knn_g, knn_grid, "KNN",                 f"Q6_{gender_label}_KNN_Tuned")

    return {"LR":lr_m, "SVM":svm_m, "KNN":knn_m}

# 6.1 Train tuned models on Male only
male_metrics = run_gender_block("Male")

# 6.2 Train tuned models on Female only
female_metrics = run_gender_block("Female")

# 6.3 Final comparison table
# Rows: SVM, KNN, LR
# Columns: Combined_A/P/R/F1, Male_A/P/R/F1, Female_A/P/R/F1

def row_for(model_key):
    # Combined metrics come from your tuned general models computed in Q4
    if model_key == "SVM":
        c = gen_svm_metrics
    elif model_key == "KNN":
        c = gen_knn_metrics
    else:
        c = gen_lr_metrics
    return {
        "Combined_A": c["Accuracy"],
        "Combined_P": c["Precision"],
        "Combined_R": c["Recall"],
        "Combined_F1": c["F1"],
        "Male_A":   male_metrics[model_key]["Accuracy"],
        "Male_P":   male_metrics[model_key]["Precision"],
        "Male_R":   male_metrics[model_key]["Recall"],
        "Male_F1":  male_metrics[model_key]["F1"],
        "Female_A": female_metrics[model_key]["Accuracy"],
        "Female_P": female_metrics[model_key]["Precision"],
        "Female_R": female_metrics[model_key]["Recall"],
        "Female_F1":female_metrics[model_key]["F1"],
    }

final_rows = []
for m in ["SVM","KNN","LR"]:
    record = {"Model": m}
    record.update(row_for(m))
    final_rows.append(record)

comparison_df = pd.DataFrame(final_rows)
print("\nFinal comparison table. Rows are SVM, KNN, LR. Columns are Combined, Male, Female with A, P, R, F1.")
display(comparison_df)

# Save table and an F1 comparison plot
comparison_df.to_csv(OUT_DIR / "Q6_Final_Comparison_Table.csv", index=False)

plt.figure(figsize=(8,5))
plot_df = pd.melt(
    comparison_df[["Model","Combined_F1","Male_F1","Female_F1"]],
    id_vars="Model", var_name="Scope", value_name="F1"
)
plot_df["Scope"] = plot_df["Scope"].str.replace("_F1","", regex=False)
sns.barplot(data=plot_df, x="Model", y="F1", hue="Scope")
plt.ylim(0,1)
plt.title("F1 comparison. Combined vs Male vs Female")
plt.tight_layout()
plt.savefig(FIG_DIR / "Q6_Fig1_F1_Comparison_Combined_Male_Female.png", dpi=200)
plt.show()

# 6.4 Optional. Learning rate test (OFF by default)
# If you need to analyze learning rate, set DO_LR_EXPERIMENT = True.
DO_LR_EXPERIMENT = False
if DO_LR_EXPERIMENT:
    from sklearn.linear_model import SGDClassifier

    def sgd_experiment(loss_name, Xtr, ytr, Xte, yte, tag):
        clf = SGDClassifier(loss=loss_name, random_state=SEED, max_iter=2000, tol=1e-3)
        grid = {
            "learning_rate":["constant","optimal","invscaling","adaptive"],
            "eta0":[0.0001, 0.001, 0.01, 0.1],
            "alpha":[1e-4, 1e-3]
        }
        gs = GridSearchCV(clf, grid, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED),
                          n_jobs=-1, scoring="accuracy", refit=True, verbose=0)
        gs.fit(Xtr, ytr)
        best = gs.best_estimator_
        y_pred = best.predict(yte)
        met = evaluate_model(yte, y_pred, f"SGD {loss_name} Learning rate tuned", f"Q6_LR_{tag}")
        return met, gs.best_params_

    sgd_log_metrics, sgd_log_best = sgd_experiment("log_loss", X_train_scaled, y_train_res, X_test_scaled, y_test, "Logistic")
    sgd_hng_metrics, sgd_hng_best = sgd_experiment("hinge",     X_train_scaled, y_train_res, X_test_scaled, y_test, "LinearSVM")

    pd.DataFrame({"SGD_Logistic": sgd_log_metrics, "SGD_LinearSVM": sgd_hng_metrics}).T.to_csv(OUT_DIR / "Q6_LearningRate_Combined_Metrics.csv")
    pd.DataFrame({"SGD_Logistic": sgg_log_best,    "SGD_LinearSVM": sgd_hng_best}).to_csv(OUT_DIR / "Q6_LearningRate_Best_Params.csv")
```
